```python
# delete Price value from calendar since its missing 83% of the values 
df_calendar.drop(columns=['price'], inplace=True)

# Verify that the column is removed
print(df_calendar.columns)
```

```python
# Fill missing 'last_review' with a placeholder
df_listings['last_review'].fillna('Never Reviewed', inplace=True)

# Fill missing 'reviews_per_month' with 0 (indicating no reviews)
df_listings['reviews_per_month'].fillna(0, inplace=True)

# For 'name', you can fill with "Unknown" if you need to keep the rows
df_listings['name'].fillna('Unknown', inplace=True)

# Check if the missing values were filled
print(df_listings.isnull().sum())
```

```python



# 1. Leave `neighbourhood` as-is (no imputation)
# No action needed for 'neighbourhood' as we leave the missing values.

# 2. Drop the `neighbourhood_group_cleansed` column if it exists
if 'neighbourhood_group_cleansed' in df_listings_details.columns:
    df_listings_details.drop('neighbourhood_group_cleansed', axis=1, inplace=True)

# 3. Leave `city` as-is (no imputation)
# No action needed for 'city' as we leave the missing values.

# 4. Leave `zipcode` and `state` columns as they are with missing values (no action needed)

# 5. Convert relevant columns to numeric (removing currency symbols and commas)
def convert_to_numeric(col):
    # Remove any non-numeric characters like '$' and ',' and convert to numeric
    return pd.to_numeric(col.replace({'\$': '', ',': ''}, regex=True), errors='coerce')

# List of columns that are supposed to be numeric but may contain non-numeric values
numeric_columns = ['price', 'weekly_price', 'monthly_price', 'security_deposit', 
                   'cleaning_fee', 'bathrooms', 'bedrooms', 'beds']

# Apply the conversion function to those columns
for col in numeric_columns:
    df_listings_details[col] = convert_to_numeric(df_listings_details[col])

# 6. Impute other missing values in columns (e.g., 'space', 'access', etc.)
# For text columns like 'space', 'access', 'neighborhood_overview' etc., fill with placeholder text
df_listings_details['space'].fillna('No space description available', inplace=True)
df_listings_details['access'].fillna('No access description available', inplace=True)
df_listings_details['neighborhood_overview'].fillna('No overview available', inplace=True)

# 7. For numerical columns, impute missing values with the median (e.g., `bathrooms`, `bedrooms`, `security_deposit`, etc.)
numerical_cols = ['bathrooms', 'bedrooms', 'beds', 'security_deposit', 'cleaning_fee', 'weekly_price', 'monthly_price']
for col in numerical_cols:
    df_listings_details[col].fillna(df_listings_details[col].median(), inplace=True)

# 8. For columns like `price`, which should have no missing values, ensure they are correctly filled (e.g., impute with median if needed)
df_listings_details['price'].fillna(df_listings_details['price'].median(), inplace=True)

# 9. For review score columns (e.g., `review_scores_rating`, `review_scores_accuracy`, etc.), we can impute with the mean or median.
review_cols = [
    'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
    'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value'
]
for col in review_cols:
    df_listings_details[col].fillna(df_listings_details[col].median(), inplace=True)

# If you're done with all the missing values, you can check the result
print(df_listings_details.isnull().sum())  # Check if there are any missing values left

# Save the cleaned data if needed
# df_listings_details.to_csv('cleaned_listing_details.csv', index=False)

```

## Checking for different mixed type data

```python
# for df_listings
for col in df_listings.columns.tolist():
    weird = (df_listings[col].apply(type) != type(df_listings[col].iloc[0])).any()
    if weird:
        print(col)
    else: 
        print('There is no mixed-type data')
```

```python
# List of columns that were flagged as potentially having mixed types
columns_to_check = ['longitude', 'price_per_night', 'reviews_per_month', 
                    'calculated_host_listings_count', 'availability_365']

for col in columns_to_check:
    # Display the types of the first few entries in each column
    print(f"Checking column: {col}")
    print(df_listings[col].apply(type).head(10))  # Show the first 10 types
    print("---" * 20)
```

```python
# Force the conversion of these columns to the correct types
df_listings['longitude'] = df_listings['longitude'].astype(float)
df_listings['price_per_night'] = df_listings['price_per_night'].astype(float)
df_listings['reviews_per_month'] = df_listings['reviews_per_month'].astype(float)
df_listings['calculated_host_listings_count'] = df_listings['calculated_host_listings_count'].astype(int)
df_listings['availability_365'] = df_listings['availability_365'].astype(int)
```
